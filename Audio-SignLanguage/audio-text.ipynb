```python
import speech_recognition as sr
def speech_text():
    r=sr.Recognizer()
    mic=sr.Microphone()
    while True:
        with mic as source:
            print("speak...")
            r.adjust_for_ambient_noise(source,duration=1)
            r.energy_threshold=700
            audio=r.listen(source)
            try:
                text=r.recognize_google(audio)
                print("you said:",text)
            except:
                print("Did't heat anything please repeat")
speech_text()
```
